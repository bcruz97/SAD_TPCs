**4.3-Analise NaN's (ou Nulls) e Outliers:**

```

#Existem 14 NaN's

#Outliers:
import numpy as np
import pandas as pd
outliers=[]
def detect_outlier(movies_dt):
    
    threshold= np.mean(movies_dt) # threshold é a média
    mean_1 = np.mean(movies_dt)
    std_1 =np.std(movies_dt)
    
    
    for y in movies_dt:
        z_score= (y - mean_1)/std_1 
        if np.abs(z_score) > threshold:
            outliers.append(y)
    return outliers
```
