```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import datasets
from sklearn.decomposition import PCA
from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import chi2
from sklearn import preprocessing

#import data to variable
data = datasets.california_housing.fetch_california_housing()

#show disorganized data
data

alldata = np.column_stack((data.data,data.target))
allfnames = np.append(data.feature_names,"Target")
dframe = pd.DataFrame(alldata, columns=allfnames)

# Aggregation
PopAggregate = dframe.groupby("Population").agg(["mean"])
print("-- Aggregation --\n")
PopAggregate
```

```
# Sampling
sample = dframe.sample(10, replace="true")
print("-- Sampling --\n")
sample
```

```
# Dimensionality Reduction 
pca = PCA(3)
pca.fit(dframe)
df_pca = pca.transform(dframe)
print("-- Dimensionality Reduction --\n")
df_pca
```

```
# Feature Subset Selection
pop = dframe.loc[dframe["Population"]<1000,:]
print("-- Feature Subset Selection --\n")
pop.head()
```

```
# Feature Creation
new_data = np.concatenate((data.data, data.target.reshape(-1,1)*1), axis=1)
names = data.feature_names + ["KidsPerHouse"]
dframe = pd.DataFrame(data=new_data, columns=names)
print("-- Feature Creation --\n")
dframe
```
